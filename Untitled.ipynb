```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

# Path to the file | Ruta al archivo

data_folder = Path(".") / "buenos_aires-master"
print()
```

```python
# Open the file | Abrimos el archivo

data = pd.read_csv(data_folder/"temperatura365d_todo.csv", sep= ";")
```

```python
data.head()
```

Como primera dificultad obtuvimos que los datos del Servicio Metereologico Nacional en formato txt, por lo que un primer desafio fue transformar
esto a csv. Luego debiamos realizar una limpieza de datos eliminando duplicados y NA. 

```python
#Transformamos en str y eliminamos los dos ultimos digitos
data['FECHA'] = data["FECHA"].astype(str)
data["FECHA"] = data["FECHA"].str.slice(stop=-2)
# para evitar posibles errores de formato, se utiliza zfill(), este codigo lo que hace es agregar 0 al inicio del dato
# hasta llegar a lenght= 8, en el caso de que el dato ya tenga 8 digitos, no hace nada.. el objetivo de esto es 
# que en el caso que existan fechas escritas como 9032023, la va a convertir a 09032023, y asi evitar futuros errores
data["FECHA"] = data["FECHA"].str.zfill(8)
```

```python
print(data)
```

```python
#Buscamos si hay NA
data.isnull().values.any()
```

```python
#eliminamos NA
data = data.dropna()
print(data)
```

```python
data.shape
```

```python
#elimino filas duplicadas
data.drop_duplicates()
```

```python
rows = []
for index, row in data.iterrows():
    if row['NOMBRE'] == 'AEROPARQUE AERO' or row['NOMBRE'] == 'BUENOS AIRES OBSERVATORIO' or row['NOMBRE'] == 'EL PALOMAR AERO' or row['NOMBRE'] == 'BUENOS AIRES' or row['NOMBRE'] == 'MORON AERO' or row['NOMBRE'] == 'MERLO AERO' or row['NOMBRE'] == 'NUEVE DE JULIO' :
        rows.append(row)
    
```

```python
new_df = pd.DataFrame(rows)
print(new_df)
```

```python
estaciones = new_df.groupby('NOMBRE')

print(estaciones)
```

```python
for name, group in estaciones:
    print(name, "\n", group)
```

```python
# redefinimos la columna FECHA asignandole un formato, en este caso le decimos que los 2 primeros digitos son dia
# los digitos 3-4 son el mes y los ultimos 4 son el año 
new_df["FECHA"]= pd.to_datetime(new_df["FECHA"], format='%d%m%Y')
```

```python
import seaborn as sns
sns.set_theme(style="darkgrid")
sns.set(rc={"figure.figsize":(15, 12)}) #width=3, #height=4
# Plot the responses for different events and regions
ax = sns.lineplot(x="FECHA", y="TMIN",
             hue="NOMBRE",
             data=new_df)
ax.set_title('Variacion de temperatura maxima en estaciones meteorologicas de BA',fontdict={'size': 20})

ax.set_xlabel('Fecha', fontdict={'size': 20})
ax.set_ylabel('Temperatura Minima °C', fontdict={'size': 20})
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))

for name, group in estaciones:
    plt.hist(
        group["TMIN"], 
        histtype='bar', 
        stacked=True, 
        rwidth=0.8, 
        alpha=0.7, 
        label=name
    )
    
plt.legend()
plt.xlabel('Temperatura Minima (°C)', size=12)
plt.show()
```
